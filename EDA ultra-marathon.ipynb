```python
import pandas as pd
import seaborn as sns

df = pd.read_csv ("EDA ultra-marathon.csv")
```

```python
df.head(10)
```

```python
df.shape
print(df.shape)

df.dtypes
```

```python
# I'm only looking at the races in US, 50km or 50mi, 2020

df[df['Event distance/length'] == '50mi']
```

```python
# Combining 50km and 50mi with isin & searching for USA events only

df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains(r'\(USA\)'))]
```

```python
df_filtered = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains(r'\(USA\)'))]

df_filtered.head(10)
```

```python
df_filtered.shape
```

```python
# Remove USA from event name

df_filtered['Event name'] = df_filtered['Event name'].str.split('(').str.get(0)
df_filtered.head()
```

```python
# Clean up athlete age 

df_filtered['Athlete age'] = 2020 - df_filtered['Athlete year of birth']
```

```python
# Removing h from athlete performance

df_filtered['Athlete performance'] = df_filtered['Athlete performance'].str.replace(' h', '')

df_filtered.head()
```

```python
# Drop columns: Athelete club, Athlete country, Athlete year of birth, Athlete age category

df_filtered = df_filtered.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)

df_filtered.head()
```

```python
# Cleaning up null values

df_filtered.isna().sum()
```

```python
df_filtered[df_filtered['Athlete age'].isna()==1]
```

```python
df_filtered = df_filtered.dropna()
df_filtered.shape
```

```python
# Check for duplicates

df_filtered[df_filtered.duplicated() == True ]
```

```python
# Resetting index

df_filtered.reset_index(drop = True)
```

```python
# Fix types

df_filtered.dtypes
```

```python
df_filtered['Athlete age'] = df_filtered['Athlete age'].astype(int)
df_filtered['Athlete average speed'] = df_filtered['Athlete average speed'].astype(float)

df_filtered.dtypes
```

```python
# Renaming the columns

df_filtered = df_filtered.rename (columns = {'Year of event': 'year',
                                             'Event dates': 'race_day',
                                             'Event name': 'race_name',
                                             'Event distance/length': 'race_lenght',
                                             'Event number of finishers': 'race_number_of_finishers',
                                             'Athlete performance': 'athlete_performance',
                                             'Athlete gender': 'athlete_gender',
                                             'Athlete average speed': 'athlete_average_speed',
                                             'Athlete ID': 'athlete_id',
                                             'Athlete age': 'athlete_age',
                                            })


df_filtered.head()
```

```python
# Reorder columns

df_sort = df_filtered[['race_day', 'race_name', 'race_lenght', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
df_sort.head()
```

```python
# Questions I want to answer with my data

# 1. Difference in speed for the 50km and 50mi between males and females

df_sort.groupby(['race_lenght', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# 2. What age groups are the best in the 50mi race with 20+ races minimum? (First 15)

df_sort.query('race_lenght == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
# 3. What age groups are the worst in the 50mi race with 10+ races min? (First 15)

df_sort.query('race_lenght == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>9').head(15)
```

```python
# 4. We are looking at different seasons -> Slower in summer than winter?

# Spring 3-5
# Summer 6-8
# Fall 9-11
# Winter 12-2

# Split between 2 decimals
```

```python
df_sort['race_month'] = df_sort['race_day'].str.split('.').str.get(1).astype(int)
df_sort.head(25)
```

```python
df_sort['race_season'] = df_sort['race_month'].apply(lambda x:'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
df_sort.head(25)
```

```python
df_sort.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

```python
# Looking at 50mi only

df_sort.query('race_lenght == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
